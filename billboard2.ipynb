```python
import pandas as pd
data = pd.read_csv('billboard.csv')
data
```

```python
data = pd.melt(data, id_vars = ['artiste', 'track','genre','date.entered','date.peaked'],var_name = 'week',value_name = 'rank')
data.fillna(0,inplace = True)
data
```

```python
data['week'] = data['week'].str.extract('(\d+)').astype(int)
data
```

```python
data['date.entered'] = pd.to_datetime(data['date.entered'])
data['date.peaked'] = pd.to_datetime(data['date.peaked'])
data['entered_to_peak'] = data['date.peaked'] - data['date.entered']
data['entered_to_peak'] = data['entered_to_peak'].dt.days
data
```
