```python
data['date.entered'] = pd.to_datetime(data['date.entered'])
data['date.peaked'] = pd.to_datetime(data['date.peaked'])
data['entered_to_peak'] = data['date.peaked'] - data['date.entered']
data['entered_to_peak'] = data['entered_to_peak'].dt.days
data
```

```python
df2 = data.set_index(['artiste', 'track','genre','date.entered','date.peaked','entered_to_peak']).rename_axis(['week'], axis=1).stack()
```
