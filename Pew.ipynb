```python
import pandas as pd
data = pd.read_csv('pew.csv')
data
```

```python
data = data.rename(columns={
    '<$10k': '10k',
    '$10-20k': '10-20k',
    '$20-30k': '20-30k',
    '$30-40k': '30-40k',
    '$40-50k': '40-50k',
    '$50-75k': '50-75k'
})
data
```

```python
data['10-201k'] = data['10k'] + data['10-20k']
data
```

```python
data = data[['religion','10-201k','20-30k','30-40k','40-50k','50-75k']]
data
```

```python
data = data.rename(columns={'10-201k': '10-20k'})
data
```

```python
data.index.name = 'data_id'
data
```

```python
df1 = pd.melt(data, id_vars = 'religion', value_vars = ['10-20k','20-30k','30-40k','40-50k','50-75k'],value_name = 'frequency', var_name = 'income_range')
df1
```
